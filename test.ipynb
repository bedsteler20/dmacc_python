```python
import random
import string
import time

def generate_words(amount):
    words = []
    for i in range(amount):
        word = ''.join(random.choices(string.ascii_lowercase, k=5))
        words.append(word)
    return words


def switch_dict(value: str):
    def case_abcd():
        return 1
    
    def case_efgh():
        return 2
    
    def case_fooo():
        return 3
    
    def case_barr():
        return 4
    
    def case_default():
        return 5
    
    switch = {
        'abcd': case_abcd,
        'efgh': case_efgh,
        'fooo': case_fooo,
        'barr': case_barr,
    }

    return switch.get(value, case_default)()


def switch_if(value: str):
    if value == 'abcd':
        return 1
    elif value == 'efgh':
        return 2
    elif value == 'fooo':
        return 3
    elif value == 'barr':
        return 4
    else:
        return 5
    
def switch_match(value: str):
    match value:
        case 'abcd':
            return 1
        case 'efgh':
            return 2
        case 'fooo':
            return 3
        case 'barr':
            return 4
        case _:
            return 5
        

def test_function(func, words, times=100):
    all_times = []
    
    for i in range(times):
        start = time.time()
        for word in words:
            func(word)
        end = time.time()
        all_times.append(end - start)
    
    print(f'{func.__name__} took {sum(all_times) / times} seconds on average')


def main():
    words = generate_words(1000000)
    test_function(switch_dict, words)
    test_function(switch_if, words)
    test_function(switch_match, words)


if __name__ == '__main__':
    main()
```
